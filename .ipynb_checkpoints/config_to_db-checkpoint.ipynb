```python
direct_det_mapping_siia = {
    1430:{
        'stp':{1:[1,2],2:[3,4,5],3:[9], 4:[10,11],5:[12], 6:[13, 14],7:[],8:[19, 20]},
        'adv':{}
    },
    
    1435:{
         'stp':{1:[1],2:[2,3,4],3:[18,19,20], 4:[18,19,20],5:[10], 6:[11,12,13],7:[],8:[]},
        'adv':{2:[5,6,7],6:[14,15,16]}
    }
    
}


exp_dict = {
    
    'inputs':{'isc':1430,
              'type':'stp',
             'phases':[2,4,8],
             'sig':True},
    
    'outputs' : {
        'isc':1435,
        'type':'adv',
        'phases' :[2],
        'sig':False}

    
}
```

```python
final_res =[]
for each_isc, val in direct_det_mapping_siia.items():
    for each_type, ty in val.items():
        for each_ph, ph in ty.items():
            for each_ch in ph:
                final_res.append({
                    'SignalID':each_isc,
                    'DetType':each_type,
                    'phase': each_ph,
                    'channel':each_ch,
                })
            
```

```python
meta_data_df = pd.DataFrame(final_res)
```

```python
# meta_data_df['channel_str'] = [f"0{x}" if x <10 else str(x) for x in meta_data_df['channel'] ]

meta_data_df['channel_str'] = [str(x) for x in meta_data_df['channel'] ]
```

```python
meta_data_df['DetectorID'] = [str(x) + y for x,y in zip(meta_data_df['SignalID'],  meta_data_df['channel_str'])]     
```

```python
meta_data_df['SignalID'] = [str(x) for x in meta_data_df['SignalID']]
```

```python
meta_data_df.drop(['channel_str'],axis = 1, inplace = True)
```
